```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '76daa7075f7dc2e4a339f93d8f71288d'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
def drget(x):
    if "(director)" in x:
        return x.split("(director)")[0].strip()
    elif "(director/screenplay)" in x:
        return x.split("(director/screenplay)")[0].strip()
    else:
        return x.split(" (director/screenplay)")[0].strip()
```

```python
df["director_name"] = df['Cast and crew'].map(lambda x : drget(x))
```

```python
def screenplaying(x):
    c = 0
    if "(screenplay)" in x:
        c+=1
    elif "(co-director/screenplay)" in x:
        c+=1
    elif "(director/screenplay)":
        c+=1
    return c
```

```python
df['Cast and crew'].map(lambda x : screenplaying(x))
```

```python
df['Cast and crew'][1].split("(screenplay);")
```

```python
def get_actor(x,num):
    try:
        if "(screenplay); " in x:
            return ((x.split("screenplay); ")[-1]).split(", ")[num])
        elif "(co-director/screenplay);" in x:
            return ((x.split("co-director/screenplay); ")[-1]).split(", ")[num])
        elif "(director/screenplay);" in x:
            return ((x.split("director/screenplay); ")[-1]).split(", ")[num])
    except:
        return np.nan
```

```python
df['actor_1_name'] = df['Cast and crew'].map(lambda x: get_actor(x,0))
```

```python
df['actor_2_name'] = df['Cast and crew'].map(lambda x: get_actor(x,1))
```

```python
df['actor_3_name'] = df['Cast and crew'].map(lambda x: get_actor(x,2))
```

```python
df.Title = df.Title.str.lower()
```

```python
df.head()
```

```python
df = df.drop(['Cast and crew'], axis = 1)
```

```python
df = df.rename(columns={"Title" : "movie_title"})
```

```python
df['actor_2_name'] = df['actor_2_name'].replace(np.nan, 'unknown')
df['actor_3_name'] = df['actor_3_name'].replace(np.nan, 'unknown')
df['actor_1_name'] = df['actor_1_name'].replace(np.nan, 'unknown')
df['genres'] = df['genres'].replace(np.nan,'unknown')
```
