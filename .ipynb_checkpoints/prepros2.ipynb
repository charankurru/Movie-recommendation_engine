```python
import ast
newdata['genres'] = newdata['genres'].map(lambda x: ast.literal_eval(x))
newdata['cast'] = newdata['cast'].map(lambda x: ast.literal_eval(x))
newdata['crew'] = newdata['crew'].map(lambda x: ast.literal_eval(x))
```

##### there could be a chance of having two director for a movie

```python
def directorCountFinder(x):
    c = 0
    for i in x:
        if i.get("job") == "Director":
            c+=1
    return c
```

```python
directorCountFinder(newdata['crew'][0])
```

#### getting director Names and with column name = "director_name"

```python
def getDirector(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
newdata['director_name'] = newdata['crew'].map(lambda x :getDirector(x))
```

```python
newdata['director_name']
```

### getting actor_names from cast column

```python
newdata.head()
```

```python
newdata['cast'][0]
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
    
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
    
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
newdata['actor_1_name'] = newdata['cast'].map(lambda x: get_actor1(x))
newdata['actor_2_name'] = newdata['cast'].map(lambda x: get_actor2(x))
newdata['actor_3_name'] = newdata['cast'].map(lambda x: get_actor3(x))
```

### converting Genres Data

```python
test['genres'][0]
```

```python
def GenresConverter(x):
    genre = []
    st = " "
    for i in x:
        if i.get("name") == "Science Fiction":
            genre.append("Sci-Fi")
        else:
            genre.append(i.get("name"))
    return (st.join(genre))
```

```python
test['genres'] = test['genres'].apply(lambda x:GenresConverter(x))
```

```python
test.head()
```

```python
final = test.loc[:,["director_name","actor_1_name","actor_2_name","actor_3_name","genres","movie_title"]]
```

```python
final.movie_title = final.movie_title.str.lower()
```

```python
final['comb'] = final['director_name']+final['actor_1_name']+final["actor_2_name"]+final["actor_3_name"]+final["genres"]+final["movie_title"]
```

```python
final = final.dropna(how='any')
```
